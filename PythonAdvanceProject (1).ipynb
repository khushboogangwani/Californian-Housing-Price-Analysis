```python
'''4. Create a data set by deleting the corresponding examples from
the data set for which total_bedrooms are not available.'''

# Assuming housing is original DataFrame
# Create a new DataFrame by removing rows with missing 'total_bedrooms'
new_dataset = housing.dropna(subset=['total_bedrooms']).copy()

# Display basic information about the new dataset
print("Basic Information about the New Dataset:")
print(new_dataset.info())
```

```python
'''5. Create a data set by filling the missing data with the mean value
 of the total_bedrooms in the original data set.'''

# Calculate the mean of the 'total_bedrooms' column
mean_total_bedrooms = housing['total_bedrooms'].mean()

# Create a new DataFrame with missing values filled by the mean
new_dataset_filled = housing.copy()
new_dataset_filled['total_bedrooms'].fillna(mean_total_bedrooms, inplace=True)

# Display basic information about the new dataset
print("Basic Information about the New Dataset with Filled Values:")
print(new_dataset_filled.info())


```

```python
'''6. Write a programming construct (create a user defined function)
to calculate the median value of the data set wherever required'''

def calculate_median(data, median_income ):
   # Check if the column_name is present in the DataFrame
    if median_income not in data.columns:
        raise ValueError(f"Column '{median_income }' not found in the dataset.")

    column_data = data[median_income ]
    median_value = column_data.median()

    return median_value

# Example usage:
housing_median = calculate_median(housing, 'median_income')
print(f"The median value for 'median_income' is: {housing_median}")

```

```python
'''10. Please create a new column named total_bedroom_size. If the total bedrooms
 is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more
  but less than 1000, it should be medium, otherwise it should be considered large.'''


# Define a function to categorize total bedrooms
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Create the new column 'total_bedroom_size'
housing['total_bedroom_size'] = housing['total_bedrooms'].apply(categorize_bedroom_size)

# Display the DataFrame with the new column
print(housing[['total_bedrooms', 'total_bedroom_size']].head())

```
